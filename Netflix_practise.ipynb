# **Time-Based Analysis **

```python
date_max = df['date_added'].value_counts().index[0]
print(date_max)
df['date_added'].fillna(date_max, inplace=True)

df['date_added'] = pd.to_datetime(df['date_added'].str.strip())
df
```

```python
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
```

```python
Movie_year_stats = df[df['type'] == "Movie"]['year_added'].value_counts().sort_index()
TVshow_year_stats = df[df['type'] == "TV Show"]['year_added'].value_counts().sort_index()
plt.figure(figsize=(12,6))
plt.title("Content over years")
sns.lineplot(x=Movie_year_stats.index,y=Movie_year_stats.values,label = 'Movies')
sns.lineplot(x=TVshow_year_stats.index,y=TVshow_year_stats.values, label = 'TV Shows')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
Movie_month_stats = df[df['type'] == "Movie"]['month_added'].value_counts().sort_index()
TVshow_month_stats = df[df['type'] == "TV Show"]['month_added'].value_counts().sort_index()
plt.figure(figsize=(12,6))
plt.title("Monthly Trends")
sns.barplot(x=Movie_month_stats.index,y=Movie_month_stats.values,label = 'Movies')
sns.barplot(x=TVshow_month_stats.index,y=TVshow_month_stats.values, label = 'TV Shows')
plt.xlabel('Month')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
df_rating = df['rating'].value_counts()
#df_cntry_top10 = df_country.head(10)
plt.figure(figsize=(12,6))
plt.title("Plot of content based on rating")
sns.barplot(x=df_rating.index,y=df_rating.values)
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

From the above graphs, most of the content produced has rating TV-MA, meaning that it is for audiance under the age of 17. With the Stats added including TV-14 coming at second position, most of the content is for audiance whose age is below 14 years.

```python
df['cast'] = df['cast'].str.split(', ')
df_cast = df.explode('cast')
cast_counts = df_cast['cast'].value_counts()
```

```python
plt.figure(figsize=(12,6))
plt.title("Top 20 Actors by Frequency")
sns.barplot(x=cast_counts.index[:20],y=cast_counts.values[:20])
plt.xlabel('Cast')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
df['listed_in'] = df['listed_in'].str.split(', ')
df_genre = df.explode('listed_in')
genre_counts = df_genre['listed_in'].value_counts()
```

```python
plt.figure(figsize=(12,6))
plt.title("Top Genres by Frequency")
sns.barplot(x=genre_counts.index[:20],y=genre_counts.values[:20])
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
df_dur_max = df['duration'].value_counts().index[0]
df['duration'].fillna(df_dur_max, inplace=True)
```

```python
def split_dur(dur_str):
  if 'Season' in dur_str:
    return int(dur_str.split(" ")[0]), 0
  else:
    return 0, int(dur_str.split(" ")[0])

df[['Seasons','Movie Duration']]= df['duration'].apply(lambda x: split_dur(str(x))).apply(pd.Series)
```

```python
yearly_minutes = df.groupby('release_year')['Movie Duration'].sum()
#yearly_TV_seasons = df.groupby('release_year')['Seasons'].sum()
plt.figure(figsize=(6,6))
plt.title("Content over years")
sns.lineplot(x=yearly_minutes.index,y=yearly_minutes.values,label = 'Movies')
#sns.lineplot(x=yearly_TV_seasons.index,y=yearly_TV_seasons.values, label = 'TV Shows')
plt.xlabel('Year')
plt.ylabel('Minutes')
plt.xticks(rotation=90)
plt.show()
```
