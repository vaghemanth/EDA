## ANOVA TEST ##

ANOVA FEASIBILITY CHECK

```python
#{1: "Spring", 2: "Summer", 3: "Fall", 4: "Winter"}
df_spring = df[df["season"] == "Spring"]["count"]
df_summer = df[df["season"] == "Summer"]["count"]
df_fall = df[df["season"] == "Fall"]["count"]
df_winter = df[df["season"] == "Winter"]["count"]

print(f"Spring Count mean: {df_spring.mean()}")
print(f"Summer Count mean: {df_summer.mean()}")
print(f"Fall Count mean: {df_fall.mean()}")
print(f"Winter Count mean: {df_winter.mean()}")
```

Visual check for Normal Distribution

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(16, 12))
sns.kdeplot(data=df_spring.reset_index(), x='count', ax=axs[0,0])
sns.kdeplot(data=df_summer.reset_index(), x='count', ax=axs[0,1])
sns.kdeplot(data=df_fall.reset_index(), x='count', ax=axs[1,0])
sns.kdeplot(data=df_winter.reset_index(), x='count', ax=axs[1,1])
plt.show()
```

```python
_, p_sprg = stats.shapiro(df_spring)
_, p_smr = stats.shapiro(df_summer)
_, p_fall = stats.shapiro(df_fall)
_, p_wntr = stats.shapiro(df_winter)

print(f"Spring prob:{p_sprg}\
      Summer prob:{p_smr} \
      Fall prob:{p_fall}\
      Winter prob:{p_wntr}")
```

From the above probability statistics for Seasons- Spring, Summer, Fall, Winter the p-value is seen to be very low for Shapiro test. So none of the Distributions are Normal which denies the first conditions for ANOVA.

```python
_, p_lvn = stats.levene(df_spring, df_summer, df_fall, df_winter)

alpha = 0.05

if(p_lvn < alpha):
  print("Reject Null hypothesis, Variance not same")
else:
  print("Fail to reject Null hypothesis, Vaiance same")
```

Another Prerequisite for ANOVA of same Variance among these Seaons distributions are not satisfied. The Levene test rejects the null hypothesis indicating different Variance. So, we go ahead with Kruskal's test.

KRUSKAL Test

```python
_, p_krus = stats.kruskal(df_spring, df_summer, df_fall, df_winter)

print(p_krus)

alpha = 0.05

if(p_krus < alpha):
  print("Reject Null hypothesis, Seasons have impact on cycle usage")
else:
  print("Fail to reject Null hypothesis, Seasons have no impact on cycle usage")
```

From the above result we conclude we dont have enough evidence that the Seasons impact the cycle usage.

```python
_, p_anova = stats.f_oneway(df_spring, df_summer, df_fall, df_winter)

print("p_value:",p_anova)

alpha = 0.05

if(p_anova < alpha):
  print("Reject Null hypothesis, Seasons have impact on cycle usage")
else:
  print("Fail to reject Null hypothesis, Seasons have no impact on cycle usage")
```

Even Though ANOVA test is not suitable for this data, we try out what could be the result. The result seems similar as Kruskals.

From Shapiro test, none of the distributions have normal distributions

```python
_, p_lvn = stats.levene(df_clr, df_mst, df_lrain)
                        #df_hrain)

alpha = 0.05

if(p_lvn < alpha):
  print("Reject Null hypothesis, Variance not same")
else:
  print("Fail to reject Null hypothesis, Variance same")
```

From Levene test, the Variance is not same, hence cant do ANOVA test.

```python
_, p_krus = stats.kruskal(df_clr, df_mst, df_lrain)
                        #df_hrain)

print(p_krus)

alpha = 0.05

if(p_krus < alpha):
  print("Reject Null hypothesis, Weather has impact on cycle usage")
else:
  print("Fail to reject Null hypothesis, Weather has no impact on cycle usage")
```

From Kruskal test, it can be concluded that we dont have enough evidence to show that cycle usage is dependent on Weather.

```python
_, p_anova = stats.f_oneway(df_clr, df_mst, df_lrain)
                        #df_hrain)

print("p_value:",p_anova)

alpha = 0.05

if(p_anova < alpha):
  print("Reject Null hypothesis, Weather has impact on cycle usage")
else:
  print("Fail to reject Null hypothesis, Weather has no impact on cycle usage")
```

ANOVA test also has the similar result as kruskals.

## CHI-SQUARE TEST ##

```python
chi_stat, p_value, _, exp_freq = stats.chi2_contingency(ex_condition_crosstab) # chi_stat, p_value, df, expected value

print("chi_stat:",chi_stat)
print("p_value:",p_value)
print("exp_freq:",exp_freq)
```

```python
alpha = 0.05

if p_value < alpha:
    print("Reject H0")
    print("Season impacts weather")
else:
    print("Fail to reject H0")
    print("Season does not impact weather")
```

```python

df_workingday = df[df["workingday"] == 1]["count"]
df_not_workingday = df[df["workingday"] == 0]["count"]

print(f"Working Day mean: {df_workingday.mean()}")
print(f"Non Working Day mean: {df_not_workingday.mean()}")
```

```python
_, p_working = stats.ttest_ind(df_workingday, df_not_workingday, alternative='greater')
alpha = 0.05 # 95% confidence

if p_working < alpha:
  print('Reject H0, , more cycles on working day compared to non-working day')
else:
  print ('Fail to Reject H0,  more cycles on working day not required compared to non-working day')
```
